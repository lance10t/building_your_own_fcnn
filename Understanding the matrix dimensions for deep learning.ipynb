```
np.random.seed(42)

x = np.random.random(size=(3,1))
w = np.random.random(size=(4,3))
print(f'Shape of x: {x.shape}')
print(f'Shape of w: {w.shape}')
```

```
z_wx = np.dot(w,x)
print(z_wx)
print(z_wx.shape)
```

We can see that this works exactly as planned in our convention above.  If the dimensions are wrong, we will get the following error messages.

If your structures are similar, you are lucky because your code just won't run.  You can image if your subsequent layer is the same size as the incoming dimensions (e.g. if w was a (3,3) shape instead, and your tweaked your codes to use x.T), your code will still run, but with wrong outputs!

```
z_xw = np.dot(x,w)
```

```
# Method 2 - If x is aligned such that each training set is in a row instead, i.e. (1,3)
# Then w CANNOT be first, else you have no way of aligning the matrices
# Thus you need to use (1,3) . (3,4) = (1,4) for output
# In this case, you need to transpose the shape of w too.
x = x.T
w = w.T
z_xw = np.dot(x,w)
print(z_xw)
print(z_xw.shape)
```
